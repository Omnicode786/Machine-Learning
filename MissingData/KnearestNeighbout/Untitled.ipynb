# 🧠 KNN Weights: Uniform vs Distance vs Callable

This document explains the difference between `uniform`, `distance`, and `callable` weights in the **K-Nearest Neighbors (KNN)** algorithm.  
It is written to be used **directly inside a Jupyter Notebook Markdown cell** or saved as a standalone `.md` file.

---

## 📌 What Does `weights` Mean in KNN?

After KNN finds the **k nearest neighbors** to a query point, it must decide **how much influence** each neighbor has on the final prediction.

This influence is controlled by the `weights` parameter.

---

## 1️⃣ `weights="uniform"`

### 📖 Concept
- Every neighbor contributes **equally**
- Distance is **not considered**
- Final prediction is based on **majority voting**

### 🧠 Intuition
All neighbors have the same importance, regardless of how close or far they are.

### 🧪 Code Example
```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# Training data
X = np.array([
    [1, 1],
    [2, 2],
    [3, 3],
    [6, 6],
    [7, 7]
])

# Class labels
y = np.array([0, 0, 0, 1, 1])

# Query point
query_point = np.array([[4, 4]])

# KNN with uniform weights
knn_uniform = KNeighborsClassifier(
    n_neighbors=3,
    weights="uniform"
)

knn_uniform.fit(X, y)
prediction_uniform = knn_uniform.predict(query_point)

prediction_uniform
````

---

## 2️⃣ `weights="distance"`

### 📖 Concept

* Neighbors closer to the query point have **more influence**
* Farther neighbors have **less influence**
* Uses inverse distance weighting

### 📐 Mathematical Idea

The weight assigned to each neighbor is approximately:

[
\text{weight} = \frac{1}{\text{distance}}
]

As distance decreases, weight increases.

### 🧠 Intuition

Closer neighbors are more trustworthy, so their votes count more.

### 🧪 Code Example

```python
knn_distance = KNeighborsClassifier(
    n_neighbors=3,
    weights="distance"
)

knn_distance.fit(X, y)
prediction_distance = knn_distance.predict(query_point)

prediction_distance
```

---

## 3️⃣ `weights=callable` (Custom Weight Function)

### 📖 Concept

* You define **your own function** to compute weights
* Sklearn passes the **distances** to this function
* The function returns an array of weights
* Provides **full control** over neighbor influence

### 🧠 Intuition

You decide exactly how much each neighbor should matter.

### 🧪 Code Example

```python
def custom_weight_function(distances):
    """
    distances: array of distances from the query point
    Custom rule:
    - Very close neighbors get strong influence
    - Others get weak influence
    """
    weights = []

    for d in distances:
        if d < 2:
            weights.append(10)  # strong vote
        else:
            weights.append(1)   # weak vote

    return np.array(weights)


knn_callable = KNeighborsClassifier(
    n_neighbors=3,
    weights=custom_weight_function
)

knn_callable.fit(X, y)
prediction_callable = knn_callable.predict(query_point)

prediction_callable
```

---

## 🔑 Comparison Table

| Weight Type | Uses Distance | Level of Control | Behavior                  |
| ----------- | ------------- | ---------------- | ------------------------- |
| `uniform`   | No            | None             | Equal voting              |
| `distance`  | Yes           | Low              | Closer neighbors dominate |
| `callable`  | Yes           | Full             | Custom logic              |

---

## 🧠 One-Line Summary

* **uniform** → all neighbors vote equally
* **distance** → closer neighbors have stronger influence
* **callable** → user-defined weighting strategy

---

## ✅ When to Use Each

* Use **uniform** when data is clean and evenly spaced
* Use **distance** when nearby points are more reliable
* Use **callable** when domain knowledge matters
