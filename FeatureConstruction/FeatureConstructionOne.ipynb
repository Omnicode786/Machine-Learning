```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')[['Age','Pclass','SibSp',
    'Parch','Survived']]
```

```python
df.head(3)
```

```python
df.dropna(inplace = True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
x = df.iloc[:,0:4]
y = df.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
np.mean(cross_val_score(LogisticRegression(), x,y, scoring = 'accuracy', cv = 20))
```

## Apply Feature Construction

```python
x['Family_size'] = x['SibSp'] + x['Parch'] + 1
```

```python
x.head()
```

```python
def Value_family(num):
    if num == 1:
        # alone
        return 0
    elif num >= 1 and num <=4: 
        # small family
        return 1
    else: 
        # large family
        return 2
        
```

```python
Value_family(4)
```

```python
x['Family_type'] = x['Family_size'].apply(Value_family)
```

```python
x.head()
```

```python
x.drop(columns = ['SibSp','Parch', 'Family_size'], inplace = True)
```

```python
x.head()
```

```python
np.mean(cross_val_score(LogisticRegression(), x,y, scoring = 'accuracy', cv = 20))
```

## Feature splitting

```python
df= pd.read_csv('train.csv')
```

```python
df['Name'].head()
```

```python
df['Title'] = df['Name'].str.split(', ', expand = True)[1].str.split('.', expand = True)[0]
```

```python
df.head()
```

```python
df['Title'].describe()
```

```python

# Convert 'Survived' to numeric, coercing errors to NaN
df['Survived'] = pd.to_numeric(df['Survived'], errors='coerce')

# Make sure 'Survived' is explicitly converted to float type
df['Survived'] = df['Survived'].astype(float)

# Now group by 'Title' and calculate mean of 'Survived', then sort
(df.groupby('Title')['Survived'].mean()).sort_values(ascending=False)
```
