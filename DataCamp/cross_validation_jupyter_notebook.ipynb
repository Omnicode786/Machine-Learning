# Cross Validation Motivation

In machine learning, there is a common pitfall when evaluating models.

If we compute **R² (R-squared)** on a single test split, the score we get **depends heavily on how the data was split**.

Because of this, the result may **not truly represent the model’s ability to generalize to unseen data**.

To solve this problem, we use **Cross Validation**.


## Idea Behind Cross Validation

Instead of splitting the dataset only once, we split it **multiple times**.

For example, in **5-Fold Cross Validation**:

1. The dataset is split into **5 equal parts (folds)**.
2. Each fold gets a chance to become the **test set**.
3. The remaining folds become the **training set**.

This process repeats **5 times**.

Each time:
- Train the model on 4 folds
- Test the model on the remaining fold

Finally we compute statistics like:

- Mean score
- Standard deviation
- Confidence intervals


## Import Required Libraries

```python
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, KFold
```

## Load Dataset

We will use the **Diabetes dataset** from sklearn.

- X = features
- y = target variable

```python
data = load_diabetes()

X = data.data
y = data.target

X.shape, y.shape
```

## Create Model

```python
model = LinearRegression()
```

## K-Fold Cross Validation

We create a **KFold object**.

Parameters:
- `n_splits` → number of folds
- `shuffle=True` → randomly shuffle data before splitting


```python
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
```

## Compute Cross Validation Scores

`cross_val_score` automatically:

1. Splits the data
2. Trains the model
3. Evaluates the model

It returns **one score per fold**.


```python
scores = cross_val_score(model, X, y, cv=kfold, scoring='r2')

scores
```

## Understanding the Output

The **length of the scores array equals the number of folds**.

For example:

- 5 folds → 5 R² scores


## Compute Statistics

We compute:

- Mean performance
- Standard deviation
- 95% percentile range


```python
mean_score = np.mean(scores)
std_score = np.std(scores)

# 95% percentile interval
lower = np.quantile(scores, 0.025)
upper = np.quantile(scores, 0.975)

print("Scores:", scores)
print("Mean R2:", mean_score)
print("Standard Deviation:", std_score)
print("95% interval:", (lower, upper))
```

## Important Note

Increasing the number of folds:

- Improves reliability of evaluation
- But **increases computation time**

Common choices:

- 5-Fold CV
- 10-Fold CV

These provide a good balance between **accuracy and computation cost**.
