## Chatgpt code to make it fast

```python
import requests as req
import pandas as pd
from concurrent.futures import ThreadPoolExecutor

def fetch_pokemon(i):
    print("fetching data")

    res = req.get(f"https://pokeapi.co/api/v2/pokemon/{i}")
    jsonRes = res.json()
    print("fetching data complete")
    return {
        "name": jsonRes.get("name", ""),
        "ability": (
            jsonRes["abilities"][0]["ability"]["name"]
            if jsonRes.get("abilities") else ""
        ),
        "version": (
            jsonRes["game_indices"][0]["version"]["name"]
            if jsonRes.get("game_indices") else ""
        )
    
    }

with ThreadPoolExecutor(max_workers=10) as executor:
    
    data = list(executor.map(fetch_pokemon, range(1, 20)))
    

df = pd.DataFrame(data)
print("Data is complete")
```
