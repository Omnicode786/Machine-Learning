```python
import pandas as pd
```

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.64 Safari/537.36'}
```

```python
webpage = requests.get('https://www.scrapethissite.com/pages/simple/', headers = headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
print(soup.prettify())
```

```python
jobs = soup.find_all("h3")
```

```python
Country = []
for i in jobs:
    print(i.text.strip())
    Country.append(i.text.strip())
```

## We can also target elements via classnames as well

```python
countryPopulation = soup.find_all("span", class_ = "country-population")
```

```python
Population = []
for i in countryPopulation:
    print(i.text.strip())
    Population.append(i.text.strip())
```

```python
countryCapital = soup.find_all("span", class_ = "country-capital")
```

```python
Capital = []
for i in countryCapital:
    print(i.text.strip())
    Capital.append(i.text.strip())
```

```python
countryArea = soup.find_all("span", class_ = "country-area")
```

```python
Area =[]
for i in countryArea:
    print(i.text.strip())
    Area.append(i.text.strip())
```

```python
d = {'Country':Country, 'Population':Population,'Captial': Capital, 'Area':Area}
```
